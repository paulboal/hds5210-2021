# Calculate the number of days in the span for each row

```python
start_dates = pd.to_datetime(data['measure_start_date'].astype(str), format='%Y-%m-%dT00:00:00.000')
end_dates = pd.to_datetime(data['measure_end_date'].astype(str),format='%Y-%m-%dT00:00:00.000')
data['start'] = start_dates
data['end'] = end_dates
```

```python
data.head()
```

```python
data['days'] = (data.end - data.start).dt.days
```

```python
data.head()
```

```python
data['days'].value_counts()
```

```python
data['denominator'].astype(int)
```

```python
data = data[data['denominator'] != 'Not Available']
```

```python
data['count'] = data['denominator'].astype(int)
```

```python
data['count'].sum()
```

```python
data_by_hospital = data.groupby('hospital_name')
```

```python
data_by_hospital.count()
```

```python
data.head()
```

```python
mo_summary = data_by_hospital.aggregate({
    'start': ['min','max'], 
    'end': ['min','max'], 
    'count': ['min','max','sum']}).reset_index()
```

```python
mo_summary.head()
```

```python
mo_summary.columns = ['Hospital','start min','start max','end min','end max','count min','count max','sum']
```

```python
mo_summary
```

```python
mo_summary['days'] = (mo_summary['end max'] - mo_summary['start min']).dt.days
```

```python
mo_summary['per_day'] = mo_summary['sum'] / mo_summary['days']
```

```python
mo_summary.head()
```

```python
mo_summary['rank'] = mo_summary['per_day'].rank()
```

```python
mo_summary.sort_values('rank', ascending=False)
```
