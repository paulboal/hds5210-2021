# PandaSQL 

```python
import pandas as pd
import pandasql as ps
```

```python
data = pd.read_csv('/data/complications.csv')
```

```python
data.head()
```

```python
sql = """
select city, count(distinct [provider_id]) as providers
from data 
where county_name = 'JEFFERSON' 
group by city
"""
```

```python
cities = ps.sqldf(sql, locals())
```

```python
cities
```
