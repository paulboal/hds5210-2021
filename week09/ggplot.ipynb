# ggplot in Python

If you're familiar with the Grammar of Graphics (ggplot) package for R, then this will come in handy for you in Python.

```python
%matplotlib inline
```

```python
from plotnine import ggplot, geom_point, aes, stat_smooth, facet_wrap
from plotnine import geom_line, geom_smooth
from plotnine.data import mtcars

(ggplot(mtcars, aes('wt', 'mpg'))
 + geom_point())
```

```python
from plotnine import ggplot, geom_point, aes, stat_smooth, facet_wrap
from plotnine import geom_line, geom_smooth
from plotnine.data import mtcars

(ggplot(mtcars, aes('wt', 'mpg', color='factor(gear)'))
 + geom_point()
 + stat_smooth(method='lm')
 + facet_wrap('~gear'))
```

```python
import pandas as pd
covid = pd.read_excel('/data/covid19.xlsx')
```

```python
(ggplot(covid, aes('Day','Cases', color='factor(GeoId)'))
 + geom_point())
```

```python
# Build our cummulative sum using groupby and aggregation
covid_sum = covid.groupby(['DateRep'])['Cases'].sum().cumsum()
```

```python
# Convert that to a DataFrame that ggplot can visualize
cs = pd.DataFrame(covid_sum)
cs['DateRep'] = cs.index
```

```python
cs
```

```python
(ggplot(cs, aes('DateRep','Cases'))
 + geom_line())
```
